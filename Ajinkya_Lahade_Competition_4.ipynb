```python
import pandas as pd
import numpy as np
```

```python
Alldata = pd.read_csv('AllData.csv')
dataframe = pd.DataFrame(Alldata)
dataframe.rename(columns=lambda x: x.replace(' ', '_'), inplace=True)
```

```python
Alldata.head()
```

```python
dataframe = dataframe.drop(['Slot','Fault_impedance','Distance_identified_by_Takagi_method','Distance_identified_by_simple_impedance_method'],axis=1)
```

```python
dataframe.head()
```

```python
#For fault line
Alldata1 = dataframe.iloc[:, 0:24]
Alldata2 = dataframe.iloc[:, 24:25]
Faultline_train = Alldata1.iloc[:-1,:]
Faultline_test = Alldata1.iloc[-1:,:]
Fault1_train = Alldata2.iloc[:-1,:]
Fault1_test = Alldata2.iloc[-1:,:]

#For fault phase
Alldata3 = dataframe.iloc[:, 0:24]
Alldata4 = dataframe.iloc[:, 25:26]
FaultPhase_train = Alldata3 .iloc[:-1, :]
Faultpahse_test = Alldata3 .iloc[-1:, :]
Fault2_train = Alldata4.iloc[:-1, :]
Fault2_test = Alldata4.iloc[-1:, :]


#for fault type
Alldata5 = dataframe.iloc[:, 0:24]
Alldata6 = dataframe.iloc[:, 26:27]
Faulttype_train = Alldata5.iloc[:-1, :]
Faulttype_test = Alldata5.iloc[-1:, :]
Fault3_train = Alldata6.iloc[:-1, :]
Fault3_test = Alldata6.iloc[-1:, :]

#for fault location 
Alldata7 = dataframe.iloc[:, 0:24]
Alldata8 = dataframe.iloc[:, 27:29]
Faultloc_train = Alldata7.iloc[:-1, :]
Faultloc_test = Alldata7.iloc[-1:, :]
Fault4_train = Alldata8.iloc[:-1, :]
Fault4_test = Alldata8.iloc[-1:, :]
```

```python

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor

classf = GradientBoostingClassifier()
classf.fit(Faultline_train, Fault1_train)
Fault_Line_Pred = classf.predict(Faultline_test)

classf = GradientBoostingClassifier()
classf.fit(FaultPhase_train, Fault2_train)
Fault_phase_Pred = classf.predict(Faultpahse_test)

classf = GradientBoostingClassifier()
classf.fit(Faulttype_train, Fault3_train)
Fault_type_Pred = classf.predict(Faulttype_test)

#Regerssion_model = GradientBoostingRegressor

Regerssion_model = LinearRegression()
Regerssion_model.fit(Faultloc_train, Fault4_train)
Fault_location = Regerssion_model.predict(Faultloc_test)
```

```python
print ('Fault Line Prediction -->', Fault_Line_Pred)
print ('Fault Phase Prediction -->', Fault_phase_Pred )
print ('Fault Type Prediction -->', Fault_type_Pred) 
print ('Fault Location Prediction -->', Fault_location)
```
